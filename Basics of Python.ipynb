### Functions 
#### arbitrary arguments and Keyworded arguments

```python
# *args
def sumall(*nums):           # accepts tuple of Multiple argument
    sum=0
    for i in nums:
        sum=sum+i
    return sum
sumall(2,3,4,5,6,6,7,8,8)
# **kwargs
def multiply(**nums):       # accept a dictionary of multiple argument
    a=nums.keys()
    product=1
    for i in a:
        product=nums[i]*product
        
        
    return product
multiply(x=3,y=4,z=6)
### RETURN PACKING and unPacking return (sum,product)
```

## Generators
**A Function is called generator if it Yeild an iterator.To convert a function into Genrator
we need to use Yeild except Return. RETURN TERMINATE THE FUNCTION YEILD WILL NOT**

```python
def give_num():
    yield 5
value=give_num()
print(value) # It will give you Iterartor Object
print(next(value))
```

```python
def sum_list(data):
    num=[]
    for pos,i in enumerate(data):
        num.append(i*pos)
    return num

sum_list([1,2,3])
```

```python
def sum_nums(data): # yeild doesn't suspend the function 
    num=[]
    for i in data:
        num = num+i
        yield num
```
