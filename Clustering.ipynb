# Kmeans Clustering 

**Cluster the Data In 'K' Groups**

### How It works...
**Select K points Randomly as Centroid of cluster in Data measure the Distance to all other points from this assign the points to cluster to which they are near**
**Find the Mean of each cluster and Move the Centroid to it**
**Keep Reapeating the Process untill centroid stop moving**

### How to Select Best 'k'

### This Can be Done By Elbow method
**Plot the Graph Between Variation in Cluster along the 'k'**


```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
;
import pandas as pd
```

```python
df=pd.read_csv('Iris.csv')
df.head()
```

```python
df['Species'].unique()
```

```python
df['Species'].replace(to_replace=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
                     ,value=[0,1,2],inplace=True)
```

```python

plt.scatter(x=df['SepalLengthCm'],y=df['PetalLengthCm'],c=df['Species'],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmmeans=KMeans(n_clusters=3)
```

```python
kmmeans.fit(df)
```

## Alots Methods and Attribute

```python
kmmeans.cluster_centers_
```
