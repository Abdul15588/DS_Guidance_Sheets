### Neurons

**An artificial neuron is a connection point in an artificial neural network**

**A Simple artificial neuron is a mathematical function Insprired by biological neurons. Usually each input is separately weighted, and the sum is passed through a non-linear function known as an activation function or transfer function.**
#### A Simple artificial Neuron Consists of 
* Weighted Sum 
* And Activation Function 

```python
import math
def sigmoid(num):                     ## And Activation Function
    return (1.0/(1+math.exp(-num)))

def activate(inp,wt):                ## Weighted Sum
    h=0 
    for i,j in zip(inp,wt):
        h=h+(i*j)
    return sigmoid(h)

```

```python
inp=[.5,.3,.2]
weight=[.4,.7,.2]
res=activate(inp,weight)

print(res)
```

## Artificial Neural Network ANN

**A neural Network is A network of Layers of Neurons. Each Layer is connected to each other with a channel of specific weights. There are thre Main layers**
* Input Layer
###### Input Layer accepts the Inputs from the outside
* Hidden Layer
###### Sum of Multiplied (wieghts) Inputs are Pass on to the hidden layer.I hidden layer they are added with a particular value assisiate with each neuron called biased and then passes to activation function with decides if this neouron will get activated or not. Activated neuron are pass on to the next layers and so on. THIS IS CALLED FORWARD PROPOGATION
* Output Layer
###### On the basis of the final Values coming from the hidden layer (Which are the Probablities) Output Layer decide the Output. If Output is wronf is Propogates the Backward and Adjuct the weights until we get the Right output. This is how ANN gets trained



### Implementing Neural Network 

```python
import numpy as np
import pandas as pd
from random import random
from sklearn.model_selection import train_test_split
import tensorflow as tf
from pydub import AudioSegment
import IPython
```

```python
X=np.array([[random()/2 for _ in range(2)] for _ in range(2000)])
y=np.array([[i[0]+i[1]]  for i in X])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### Buulding Model

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(5, input_dim=2, activation='sigmoid'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

### Compliling model

```python
optimiser=tf.keras.optimizers.SGD(learning_rate=0.1)
model.compile(optimizer=optimiser,loss="MSE")
```

### Training Model

```python
model.fit(X_train,y_train,epochs=20)
```

### Evaluating Model

```python
model.evaluate(X_test,y_test,verbose=1) ## Loss is slightly higher than training data
```

### Make Predictions 

```python
data = np.array([[0.1,0.2],[0.2,0.2]])
```
