```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
import seaborn as sns;
%matplotlib inline 
;
```

```python
from sklearn.datasets import load_breast_cancer
```

###### JSON or Dictionaries (with 'Label','Data' , and other information) of Datasets

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
data=cancer['data']
col=cancer['feature_names']
target_col=cancer['target']
```

```python
df=pd.DataFrame(data=data,columns=col)
df.head()
```

```python
df2=pd.DataFrame(data=target_col,columns=['Target'])
```

```python
df=pd.concat([df,df2],axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['Target'],axis=1)
y=cancer['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

**Although the accuracy of our model is so good But there are some methods to Tune your models and its always welcoming to tune the parametrs of your model to find the Best one with best accuracy**
ONE OF THE BEST METHOD TO TUNE YOUR MODEL IS
## GRID SEARCH

```python
from sklearn.model_selection import GridSearchCV
```

#### Dict. of Parameters to tune your model 

```python
param={'C':[0.001,0.01,0.1,1,10,100,1000],
      'gamma':[0.0001,0.001,0.01,0.1,1,10,100,1000]}
```

```python
gcs=GridSearchCV(SVC(),param,verbose=1)### Print Data While Tuning Greater the number greater the Data will be printed
```

```python
gcs.fit(X_train,y_train)
```

```python
pred2=gcs.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred2))
print('\n')
print(classification_report(y_test,pred2))
```
